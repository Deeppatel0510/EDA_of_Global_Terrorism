### Variables Description



The "Final_data" variable consists of a subset of the original data, containing the following columns:

* **Year**    : The year in which the attack occurred.
* **Month**   : The month in which the attack occurred.
* **Day**     : The day of the month on which the attack occurred.
* **Country** : The country in which the attack occurred.
* **City**    : The city in which the attack occurred.
* **Region**  : The region in which the attack occurred.
* **Attack_Type**: The type of attack that occurred.
* **Target**  : The target of the attack.
* **Killed**  : The number of people killed in the attack.
* **Wounded** : The number of people wounded in the attack.
* **Summary** : A brief summary of the attack.
* **Group**   : The group responsible for the attack, if known.
* **Target_type**: The type of target that was attacked, such as a civilian, military, or government target.
* **Weapon_type** : The type of weapon that was used in the attack.
* **Success_rate**: The percentage of the attack that was successful, such as the percentage of people killed or wounded.

This subset of data focuses on specific variables related to the year, month, day, location, attack details, casualties, summary, responsible group, target type, weapon type, and success rate. Analyzing this data can provide insights into patterns, trends, and characteristics of terrorist incidents.

#   Chart - 8 : Analyzing the Top 15 Countries with the Highest Number of Killed Individuals

```python
# Group the data by 'Country' and count the number of occurrences of 'Killed'
Kill_Country = Final_data.groupby('Country')['Killed'].sum()

# Select the top 15 countries with the highest number of occurrences
Top_Country = Kill_Country.sort_values(ascending=False).head(15)

# Define a color palette
colors = sns.color_palette('Blues_r', len(Top_Country))

# Plot the count plot with different colors
plt.figure(figsize=(15,8))
sns.barplot(x=Top_Country.index, y=Top_Country.values, palette=colors)
plt.xlabel('Country')
plt.ylabel('Number of Killed Individuals')
plt.title('Top 15 Countries with the Highest Number of Killed Individuals')
plt.xticks(rotation=90)

# Add annotations to the bars
for i, value in enumerate(Top_Country.values):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.show()
```

# Chart - 10 : Analysis of Top 15 Countries by Number of Attacks and Total Kills





```python
# Select the top 15 countries by number of attacks
attack = Final_data['Country'].value_counts().head(15).to_frame()

# Group the data by country and calculate the sum of kills, then select the top 15 countries
kill = Final_data.groupby('Country')['Killed'].sum().to_frame()

# Merge the attack and kill data based on the country index
merged_data = attack.merge(kill, left_index=True, right_index=True, how='left')

# Create a bar plot
ax = merged_data.plot.bar(width=0.9)

# Set the figure size
fig = plt.gcf()
fig.set_size_inches(16, 4)

# Set the title and labels for the plot
plt.title('Top 15 Countries by Number of Attacks and Total Kills', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Display the plot
plt.show()
```

##### 1. Why did you pick the specific chart?

 I picked this specific chart to compare the number of attacks and the total number of kills in the top 15 countries, providing a comprehensive overview of their impact.

##### 2. What is/are the insight(s) found from the chart?

##  Chart - 12 : Analyzing Attack Frequency Distribution of Years in India

```python
# Number of Attacks in India by Year
india = Final_data[Final_data['Country']== 'India']
number_of_attacks = india['Year'].value_counts()
number_of_attacks
```

```python
# Set plot style
sns.set(style="whitegrid")

# Set figure size
fig, ax = plt.subplots(figsize=(15, 8))

# Create stem plot
x = india['Year'].unique()  # Unique years
y = india['Year'].value_counts().sort_index()  # Count of attacks per year
markerline, stemlines, baseline = ax.stem(x, y, linefmt='b-', markerfmt='bo', basefmt='r-')

# Adjust stem plot markers
markerline.set_markerfacecolor('b')
markerline.set_markeredgecolor('b')

# Rotate x-axis labels
plt.xticks(rotation=90)

# Set axis labels and title
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Attack Frequency Distribution of Years in India', fontsize=14)

# Remove spines
sns.despine()

# Show the plot
plt.show()
```

##### 1. Why did you pick the specific chart?

I selected this specific chart to examine the distribution of attack frequencies across different years in India.

##### 2. What is/are the insight(s) found from the chart?
